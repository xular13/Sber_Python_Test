```python
import numpy as np
```

```python
gas = np.array([1,2,3,4,5])
cost = np.array([3,4,5,1,2])
```

```python
def possible_route(gas, cost):
    num_station = len(gas)
    list_of_roots = []
    if num_station!= len(cost):
        print('Error with data')
        return -1
    for i in range(num_station):
        fuel_tank = 0
        steps = 0
        fuel_tank += gas[i]
        k = i
        while (True):
            fuel_tank -= cost[k] 
            if (fuel_tank >= 0):
                steps += 1
            else: break
            if (steps == num_station):
                list_of_roots.append(i)
                break
            fuel_tank += gas[k]
            k = (k+1)%num_station
    print('Возможные маршруты:')
    print(list_of_roots)
    return list_of_roots
```

```python
lst = possible_route(gas,cost)
```
